```python
inputs =df.drop('Salary_more_than_100k',axis='columns')
inputs
```

```python
target= df['Salary_more_than_100k']
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
inputs['company_encoded']=le_company.fit_transform(inputs['Company'])
inputs['job_encoded']=le_job.fit_transform(inputs['Job'])
inputs['degree_encoded']=le_degree.fit_transform(inputs['Degree'])
inputs
```

```python
from sklearn import tree
```

```python
input_n=inputs.drop(['Company','Job','Degree'],axis='columns')
input_n
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(input_n,target)
```
